```python
from control import tf, c2d, bode_plot
import numpy as np
import matplotlib.pyplot as plt

fc = 1500                
fs = 48000               
wc =fc     

s = tf('s')
G = wc**2 / (s**2 + np.sqrt(2) * wc * s + wc**2)

 
T = 1 / fs               
Gd = c2d(G, T)           
print(Gd)
bode_plot(Gd, dB=True, title='Bode plot de Gd', grid=False)   
plt.show()
```
